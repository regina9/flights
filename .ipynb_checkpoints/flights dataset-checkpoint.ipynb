```python
import pandas as pd
import numpy as np
flights = pd.read_csv('C:/Users/regi_/Downloads/flights.csv') #remove later
```

```python
flights.head()
```

```python
flights.isna().sum()
```

```python
columns = ["AIR_SYSTEM_DELAY", "SECURITY_DELAY", "AIRLINE_DELAY", "LATE_AIRCRAFT_DELAY", "WEATHER_DELAY", "DEPARTURE_DELAY", "ELAPSED_TIME", "AIR_TIME"]
flights[columns] = flights[columns].fillna(0)
```

```python
flights["CANCELLATION_REASON"] = flights["CANCELLATION_REASON"].fillna("N")
```

```python
flights = flights.drop(["YEAR", "TAIL_NUMBER", "WHEELS_ON", "WHEELS_OFF"], axis=1)
```

```python
flights = flights[np.isfinite(flights["SCHEDULED_TIME"])]
```

```python
cols = ["DEPARTURE_TIME", "TAXI_OUT", "TAXI_IN", "ARRIVAL_TIME", "ARRIVAL_DELAY"]
flights[cols] = flights[cols].fillna(0)
```

```python
def to_time(x):
    x = str(x)
    k = x.split(".")[0]
    hour = k[:-2]
    if hour == '':
        hour = '00'
    minute = k[-2:]
    second = '00'
    #second = x.split(".")[1]
    time = "{:02.0f}:{:02.0f}:{:02.0f}".format(int(hour), int(minute), int(second))
    return time
```

```python
flights["SCHEDULED_DEPARTURE"] = flights["SCHEDULED_DEPARTURE"].astype('float')
flights["SCHEDULED_ARRIVAL"] = flights["SCHEDULED_ARRIVAL"].astype('float')
flights["SCHEDULED_ARRIVAL"] = flights["SCHEDULED_ARRIVAL"].apply(to_time)
flights["SCHEDULED_DEPARTURE"] = flights["SCHEDULED_DEPARTURE"].apply(to_time)
```

```python
flights["DAY_OF_WEEK"] = flights["DAY_OF_WEEK"].astype('str') #convert into categorical 
```

```python
flights["ARRIVAL_TIME"] = flights["ARRIVAL_TIME"].apply(to_time)
flights["DEPARTURE_TIME"] = flights["DEPARTURE_TIME"].apply(to_time)
```
